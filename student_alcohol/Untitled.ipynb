## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Creating the scatter plot

```python
student_data = pd.read_csv('archive/student-por.csv')
```

```python
student_data.head()
```

```python
sns.scatterplot(x='absences', y = 'G3', data= student_data, hue='address')
plt.show()
```

### Number of students in urban vs rural areas, bassed on the school they go to as well

```python
# Number of students in urban vs rural areas, bassed on the school they go to as well
pallete_colors ={'R':'green','U':'Pink'}
sns.countplot(x='school', hue = 'address', data= student_data, palette = pallete_colors)
plt.show()
```

```python
sns.scatterplot(x='absences', y = 'G3', data= student_data, hue='studytime')
plt.show()
```

```python
# this shows the relation between absences and study time
sns.relplot(x='absences', y = 'G3', data= student_data, kind= 'scatter', col='studytime')
plt.show()
```

```python
sns.relplot(x='G1', y = 'G3', data= student_data, kind= 'scatter', col='schoolsup', row_order=['yes','no'])
plt.show()
```

```python
sns.relplot(x='G1', y = 'G3', data= student_data, kind= 'scatter', col='schoolsup',col_order=['yes','no'], row='famsup', row_order=['yes','no'])
plt.show()
```
